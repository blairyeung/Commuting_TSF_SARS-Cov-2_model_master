```python
import math
import os

import pandas as pd
from scipy.stats import gamma
from scipy.stats import lognorm
from scipy.stats import norm
from Util import normalize
import matplotlib.pyplot as plt
import numpy as np
import Parameters
import datetime
import cv2
```

```python
raw = np.linspace(0, 2499, 2500)
print(raw, raw.shape)
```

```python
shift = (Parameters.OUTBREAK_FIRST_DAY - datetime.datetime(2020, 1, 1)).days
print(shift)
```

```python
raw = (raw) * 2 * math.pi / 360
print(raw)
```

```python
plt.figure(figsize=(15, 4))
kernel = np.cos(raw)[shift:shift+2000]
x = [Parameters.OUTBREAK_FIRST_DAY + datetime.timedelta(days=j) for j in range(2000)]
plt.plot(x, kernel)
plt.show()
```

```python
factor = 0.2
rslt = kernel * factor + np.ones(shape=kernel.shape)
```

```python
plt.figure(figsize=(15, 4))
plt.plot(x, rslt)
plt.show()
```
